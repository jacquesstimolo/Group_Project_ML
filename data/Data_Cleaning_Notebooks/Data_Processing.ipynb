```python
import pandas as pd
```

```python
df = pd.read_csv('Cleaned_Data.csv')
df
```

```python
df.info()
```

```python
df['Sport'].nunique()
```

```python
df['NOC'].nunique()
```

```python
df = df.drop(columns=['Unnamed: 0', 'Name'], axis=1)
df
```

```python
df1 = df.groupby('Sport').ID.count().reset_index()
df1 = df1.sort_values(by='ID', ascending=False)
df1 = df1.head()
df1
```

```python
popular_sports = list(df1['Sport'].values)

df2 = pd.DataFrame()

for sport in popular_sports:
    df_aux = df[df['Sport'] == sport]
    df2 = pd.concat([df2, df_aux])
df2
```

```python
df = df2
```

```python
df = pd.concat([df, pd.get_dummies(df['NOC'])], axis=1)
df = pd.concat([df, pd.get_dummies(df['Sport'])], axis=1)

df
```

```python
df = df.drop(columns=['ID', 'NOC', 'Sport'], axis=1)
df
```

```python
df.to_csv('Processed_Data.csv')
```
