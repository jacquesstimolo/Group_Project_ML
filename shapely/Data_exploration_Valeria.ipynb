```python

clean_data = df[df['Weight'].notna()].groupby(['Sport', 'Sex', 'Height', 'Age']).Weight.median().reset_index()
```

```python
clean_data.head()
```

```python


for id in df[df['Weight'].isna()].ID:
  person = df[df['ID'] == id]
  weight = clean_data[(clean_data['Sport'] == person.iloc[0, 0]) & (clean_data['Sex'] == person.iloc[0, 1]) & (clean_data['Height'] == person.iloc[0, 2]) & (clean_data['Age'] == person.iloc[0, 3])]['Weight']
  print(weight)
  df[df['ID'] == id]['Weight'] = weight
```
