```python
df['Sex'] = pd.Series(df['Sex'], dtype="string")

def dummy_sex(series):
    res = []
    for elem in series:
        if elem == 'M':
            res.append(0)
        else:
            res.append(1)
    return res


df['Sex'] = dummy_sex(df['Sex'])
df.Sex.unique()
```

```python
def isNaN(num):
    return num != num
```

> Height and Weight have missing values\
> Make Distinction of Gender!

```python
sports = []
height_F = []
height_M = []
weight_F = []
weight_M = []

counters = []

for key_sport in df.Sport.unique():
    sports.append(key_sport)
    counter = 0
    for key in ['Height', 'Weight']:
        for i in range(2):

            count = 0
            not_NaN_heights = []
            for elem in df[(df['Sport'] == key_sport) & (df['Sex'] == i)][key]:
                if isNaN(elem):
                    continue
                else:
                    count += 1
                    not_NaN_heights.append(elem)


            if not_NaN_heights == []:
                median_height = 'missing'
            else:
                median_height = np.median(not_NaN_heights)


            if key == 'Height':
                if i == 0:
                    height_M.append(median_height)
                if i == 1:
                    height_F.append(median_height)
            if key == 'Weight':
                if i == 0:
                    weight_M.append(median_height)
                if i == 1:
                    weight_F.append(median_height)

            counter += count
            # print(key_sport, key+':', (i, median_height), count)
    counters.append(counter)

df1 = pd.DataFrame({
    'Sport' : sports,
    'Height_M' : height_M,
    'Height_F' : height_F,
    'Weight_M' : weight_M,
    'Weight_F' : weight_F,
    'Nr_samples' : counters
})

df1 = df1.sort_values(by='Nr_samples', ascending=False)
df1.reset_index(drop=True)
df1
```

```python
median_ages_m = []
median_ages_w = []

for key_sport in df.Sport.unique():
    for i in range(2):

        lis = []
        for elem in df[(df['Sport'] == key_sport) & (df['Sex'] == i)]['Age']:
            if isNaN(elem):
                continue
            else:
                lis.append(elem)
        
        if lis == []:
            if i == 0:
                median_ages_m.append('missing')
            else:
                median_ages_w.append('missing')
        else:
            if i == 0:
                median_ages_m.append(np.median(lis))
            else:
                median_ages_w.append(np.median(lis))


df1['Age_M'] = median_ages_m
df1['Age_W'] = median_ages_w
df1            
```

```python
def restore_height(sports, genders, heights):
    elements = zip(sports, genders, heights)

    res_heights = []
    for elem in elements:
        s, g, h = elem

        row_df1 = df1[df1['Sport'] == s]

        if isNaN(h):
            if g == 0:
                val = row_df1.Height_M.values
                if val == 'missing':
                    res_heights.append(median_height_m)
                else:                    
                    res_heights.append(int(val))
            else:
                val = row_df1.Height_M.values
                if val == 'missing':
                    res_heights.append(median_height_w)
                else:
                    res_heights.append(int(val))

        else:
            res_heights.append(h)
    
    return res_heights



df['Height'] = restore_height(df['Sport'], df['Sex'], df['Height'])
df['Height'] = df['Height'].astype('int64')
df
```

```python
df.info()
```

```python
def restore_weight(sports, genders, weights):
    elements = zip(sports, genders, weights)

    res_weights = []
    for elem in elements:
        s, g, w = elem

        row_df1 = df1[df1['Sport'] == s]

        if isNaN(w):
            if g == 0:
                val = row_df1.Weight_M.values
                if val == 'missing':
                    res_weights.append(median_weight_m)
                else:                    
                    res_weights.append(int(val))
            else:
                val = row_df1.Weight_M.values
                if val == 'missing':
                    res_weights.append(median_weight_w)
                else:
                    res_weights.append(int(val))

        else:
            res_weights.append(w)
    
    return res_weights


df['Weight'] = restore_weight(df['Sport'], df['Sex'], df['Weight'])
df['Weight'] = df['Weight'].astype('int64')
df
```

```python
df.info()
```

```python
def restore_age(sports, genders, ages):
    elements = zip(sports, genders, ages)

    res_ages = []
    for elem in elements:
        s, g, a = elem

        row_df1 = df1[df1['Sport'] == s]

        if isNaN(a):
            if g == 0:
                val = row_df1.Age_M.values
                if val == 'missing':
                    res_ages.append(median_age_m)
                else:                    
                    res_ages.append(int(val))
            else:
                val = row_df1.Age_M.values
                if val == 'missing':
                    res_ages.append(median_age_w)
                else:
                    res_ages.append(int(val))

        else:
            res_ages.append(a)
    
    return res_ages


df['Age'] = restore_age(df['Sport'], df['Sex'], df['Age'])
df['Age'] = df['Age'].astype('int64')
```

```python
def winner_loser(series):
    res = []
    for elem in series:
        if isNaN(elem):
            res.append(0)
        else:
            res.append(1)
    return res

df['Winner'] = winner_loser(df['Medal'])
```
