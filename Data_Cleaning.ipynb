> Dummy for Sex:\
> 0: Male, 1: Female

```python
df['Sex'] = pd.Series(df['Sex'], dtype="string")

def dummy_sex(series):
    res = []
    for elem in series:
        if elem == 'M':
            res.append(0)
        else:
            res.append(1)
    return res


df['Sex'] = dummy_sex(df['Sex'])
df.Sex.unique()
```

```python
def isNaN(num):
    return num != num
```

> Height and Weight have missing values\
> Make Distinction of Gender!

```python


for key_sport in df.Sport.unique():
    for key in ['Height', 'Weight']:
        for i in range(2):
            count = 0
            not_NaN_heights = []
            for elem in df[(df['Sport'] == key_sport) & (df['Sex'] == i)][key]:
                if isNaN(elem):
                    continue
                else:
                    count += 1
                    not_NaN_heights.append(elem)

            if not_NaN_heights == []:
                median_height = 'missing'
            else:
                median_height = np.median(not_NaN_heights)
            print(key_sport, key+':', (i, median_height), count)
```

```python
np.median([])
```

```python
def substitute_NaN_heights(series):
    res = []
    for elem in series:
        if isNaN(elem):
            res.append(median_height)
        else:
            res.append(elem)
    return res
```

```python
df['Height'] = substitute_NaN_heights(df['Height'])
df
```
